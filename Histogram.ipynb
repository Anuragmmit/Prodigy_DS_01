```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
ages = [25,30,35,40,45,50,55,60,70,75,80]
```

```python
# Creating a histogram
```

```python
plt.hist(ages, bins = 5, edgecolor = "black", alpha = 0.65)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age distribution Histogram")
plt.show()
```

```python
genders = ["Male", "Female", "Male", "Female", "Male", "Female", "Other", "Male"]
```

```python
plt.hist(genders, bins = 3, edgecolor = "black",rwidth = 0.8)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution Histogram")
plt.show()
```
